```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
unemp=pd.read_csv('Unemployment in India.csv')
unemp.head()
```

```python
 unemp.describe()
```

```python
unemp.isnull().sum()
```

```python
unemp.columns= ["Region","Date","Frequency",
               "Estimated Unemployment Rate","Estimated Employed",
               "Estimated Labour Participation Rate","Area"]
plt.title("Indian employment")
sns.histplot(x="Estimated Employed", hue="Region", data=unemp)
plt.show()
```

```python
unemp.columns=  ["Region","Date","Frequency",
               "Estimated Unemployment Rate","Estimated Employed",
               "Estimated Labour Participation Rate","Area"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=unemp)
plt.show()
```
